```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 100)
RFModNoise=(6e-5*(1+(7e-2/freq)))**2
plt.style.use('bmh')
plt.plot(freq, RFModNoise, 'b-',label = "PSD")
plt.yscale("log")
plt.xscale("log")
#plt.title("RF Modulation Noise PSD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("PSD (rad/Hz)")
plt.legend(title = "RF Modulation Noise")
plt.show()
```
